```python
import requests
import pprint
from getpass import getpass
```

```python
# APIキーの入力を受け付ける
api_key = getpass('APIキーを入力する：')
```

```python
# input関数を使って、駅名の入力を受け付ける
search_word = input('検索キーワードを入力す：')
```

```python
# 入力された駅の緯度経度を取得する
url = 'https://maps.googleapis.com/maps/api/geocode/json'

# データを取得
response = requests.get(
    url,
    params={
        'key': api_key, # APIキーをパラメータに追加
        'address': search_word
        }
)
```

```python
# # レスポンスのJSONを取得
json_data = response.json()

pprint.pprint(json_data)
```

```python
url = 'https://maps.googleapis.com/maps/api/place/nearbysearch/json'

# データを取得
response2 = requests.get(
    url,
    params={
        'location': f'{35.640128},{139.599087}',
        'radius': 500,
        'type': 'restaurant',
        'key': api_key,
     }
)
```

```python
# レスポンスのJSONを取得
json_data2 = response2.json()

# JSONの中身を確認
pprint.pprint(json_data2)
```

```python
# レストラン名、評価、住所の取得
for data in json_data2['results']:
  restaurant_name = data['name']
  evaluation = data['rating']
  address = data['vicinity']
  print(f'{restaurant_name}: {evaluation}: {address}')
```
